## merge() 중간고사 예상 문제

```python
df1 = pd.DataFrame([[1, 'apple'], [2, 'banana']], columns = ['id', 'fruits'])
df1
```

```python
df2 = pd.DataFrame([[1, 3000], [5, 500]],
                  columns = ['id', 'price'])
df2
```

```python
pd.merge(df1, df2)
```

```python
df1.merge(df2, on = 'id', how = 'inner')
```

```python
df1.merge(df2, on = 'id', how = 'left')
```

```python
df1.merge(df2, on = 'id', how = 'right')
```

```python
df1.merge(df2, how = 'cross')
```
