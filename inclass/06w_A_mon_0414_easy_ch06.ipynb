## df.agg() : 다양한 집계 연산을 수행

```python
exam = pd.read_csv('data/exam.csv')
exam.head()
```

```python
exam.agg("mean")
```

```python
exam.mean()
```

```python
exam.agg(['mean', 'median'])
```

```python
exam.agg(['mean'])
```

```python
exam.agg(['mean', 'min', 'max', 'median'])
```

```python
exam.info()
```

```python
exam.describe()
```

```python
exam.agg(['mean', 'min', 'max', 'std', 'median','count'])
```

```python
exam.agg(mean_math = ('math', 'mean'))
```

```python
exam.agg(['mean', 'median'], axis=0)
```

```python
exam.agg(['mean', 'median'], axis=1)
```

```python
exam[['math', 'english', 'science']].agg(['mean', 'median'], axis=1)
```
