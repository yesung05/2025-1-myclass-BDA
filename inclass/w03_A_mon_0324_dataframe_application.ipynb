## 데이터프레임 수정

```python
df[['사망(명)','부상(명)']].sum(axis=1) #사망+부상자 -> 행 방향 합
```

```python
df.iloc[:,1:].sum(axis=1)
```

```python
df2 = df.copy()
df2['재해자수'] = df[['사망(명)','부상(명)']].sum(axis=1)
df2
```

```python
df2.loc['총계'] = df2.sum()
df2
```

```python
df2 > 10
```

```python
df2[df2 > 10]
```
